```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as ply
import tensorflow
```

```python
# Lets do prediction
# Load the trained model & pickle file
from tensorflow.keras.models import load_model
import pickle

model=load_model('model.h5')

with open('label_encoder_gender.pkl','rb') as file:
    label_encoder_gender=pickle.load(file)

with open('onehot_encoder_geo.pkl','rb') as file:
    onehot_encoder_geo=pickle.load(file)
```

```python
test_data={
    'CreditScore':0,
    'Gender': 'Female',
    'Age':34,
    'Tenure':2, 
    'Balance':10, 
    'NumOfProducts':0,
    'HasCrCard':1, 
    'IsActiveMember':1, 
    'EstimatedSalary':100, 
    'Geography': 'France'
}
test_df=pd.DataFrame([test_data])
```

```python
test_df
```

```python
test_df['Gender']=label_encoder_gender.transform(test_df['Gender'])
geo_df=onehot_encoder_geo.transform(test_df[['Geography']]).toarray()
test_df=pd.concat([test_df,pd.DataFrame(geo_df,columns=onehot_encoder_geo.get_feature_names_out())],axis=1).drop('Geography',axis=1)
```

```python
with open('scaler.pkl','rb') as file:
    scaler=pickle.load(file)

test_df=scaler.transform(test_df)
```

```python
test_pred=model.predict(test_df)
```

```python
pred_prob=test_pred[0][0]

if pred_prob>0.5:
    print("Customer is likely to be churn")
else:
    print("Customer is not likely to churn")
```
